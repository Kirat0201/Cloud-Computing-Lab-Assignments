```python
## Multithreading and Multiprocessing

import threading
import time

def print_numbers():
  for i in range(1, 11):
    print(threading.current_thread().name, i)
    time.sleep(1)

t1 = threading.Thread(target=print_numbers, name="thread-1")
t2 = threading.Thread(target=print_numbers, name="thread-2")


t1.start()
t2.start()

t1.join()
t2.join()
```

```python
from multiprocessing import Pool

def square_sum(nums):
  return sum(i*i for i in nums)

if __name__ == "__main__":
  ranges = [
      range(1,26),
      range(26, 51),
      range(51, 76),
      range(76, 101)
  ]

  with Pool(4) as p:
    results = p.map(square_sum, ranges)

    print("Final sum:", sum (results))
```
